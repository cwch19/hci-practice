```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# <span style='color:blue'>顯示原始圖片</span>

```python
img = cv2.imread('data/face.jpg')
imgDisp = img[:,:,::-1]
# img[:,:,::-1] is numpy slicing, 
#    which means that we select all the elements on the first axis, 
#    all the elements on the seconds, and then the reverse of the elements on the third axis.

# Matplotlib expects img in RGB format but OpenCV provides it in BGR.
plt.imshow(imgDisp)
```

# <span style='color:blue'>縮小圖片</span>

```python
resImg = cv2.resize(img, (300, 200), interpolation=cv2.INTER_CUBIC)
imgDisp = resImg[:,:,::-1]
plt.imshow(imgDisp)
```

# <span style='color:blue'>圖片改灰階</span>

```python
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgDisp = grayImg[:,:]
plt.gray()
plt.imshow(imgDisp)
```

# <span style='color:blue'>圖片改亮度</span>

```python
resImg2 = np.uint8(np.clip(1.5*grayImg+10, 0, 255))
imgDisp = resImg2[:,:]
plt.imshow(imgDisp)
```

# <span style='color:blue'>裁切圖片</span>

```python
x = 515
y = 330
w = 100
h = 100

cropImg = grayImg[y:y+h, x:x+w]
imgDisp = cropImg[:,:]
plt.imshow(imgDisp)
```
